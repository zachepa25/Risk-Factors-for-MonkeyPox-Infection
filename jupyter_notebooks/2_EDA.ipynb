Visualise distribution of all variables in the dataset and print their value counts as percentages

```python
for column in df.columns: # Iterate through each column
    # Count plot
    plt.figure(figsize=(8,5))
    sns.countplot(data=df, x=column, palette='Set2')
    plt.title(f'Count of {column}')
    plt.xlabel(column)
    plt.ylabel('Count')
    plt.show()
    
    # Distribution
    counts = df[column].value_counts(normalize=True) * 100 # Get percentage distribution 
    print(f"{column} Distribution (%):")
    print(counts)
    print("\n")
```

Using Matplotlib, I created a Pie Chart to show distribution status of Mpox Test Results

```python
# Create a pie chart for Mpox Status
plt.figure(figsize=(6,6))
status_counts = df['MonkeyPox'].value_counts()
plt.pie(status_counts, labels=status_counts.index, autopct='%1.1f%%', colors=sns.color_palette('Set2'), startangle=140)
plt.title('Mpox Status Distribution')
plt.show()
```

# Section 3: Bivariate analysis

* In bivariate analysis, two variables are analyzed together to identify patterns, dependencies or interactions between them. This method helps in understanding how changes in one variable might affect another.
* I visualised the relationships by plotting various plot for the data which will show how the variables interact with each other across multiple dimensions.
* Explored associations using chi-square test of independence: A chi-square test of independence is a statistical hypothesis test used to determine if there is a significant relationship between two categorical variables. It tests whether the observed frequencies in a contingency table differ from what would be expected if the variables were independent. 
* The main interest here was the link between Mpox diagnosis and other variables

Observations

- Exploring associations between Mpox diagnosis and clinical features
    * Mpox diagnosis is only significantly associated with Swollen Tonsils

Plot MPox by each feature to see relationships and patterns. Also measured association through chi-square test of independence.
* The Chi-Square Test of Independence is a statistical test used to determine whether two categorical variables are related or independent. In other words, it checks if the distribution of one variable depends on the distribution of another. 

```python
for column in df.columns: # Iterate through each column and analyze relationship with MonkeyPox
    # Count plot with hue for MonkeyPox status
    if column != 'MonkeyPox':
        plt.figure(figsize=(8,5))
        sns.countplot(data=df, x=column, hue='MonkeyPox', palette='Set3')
        plt.title(f'Mpox Test Result by {column}')
        plt.xlabel(column)
        plt.ylabel('Count')
        plt.show()
        
    # Add chi-square test for categorical variables for boolean analysis
    from scipy.stats import chi2_contingency
    if column != 'MonkeyPox':
        contingency_table = pd.crosstab(df[column], df['MonkeyPox'])
        chi2, p, dof, expected = chi2_contingency(contingency_table, correction=True)
        print(f"Chi-square test for {column} vs Mpox:")
        print(f"Chi2 Statistic: {chi2}, p-value: {p}\n")   
    '''
        if p < 0.05:
            print(f"There is significant association between {column} and MonkeyPox (p < 0.05)\n")
        else:
            print(f"There is no significant association between {column} and MonkeyPox (p >= 0.05)\n")
    ''' 
    
```
